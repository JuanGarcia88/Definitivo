```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
import cmath
```

```python
Fe = [21, 40, 72]
Ve = 380
if Fe[-1] >= 60:
    Ve_m = Ve / 60
else:
    Ve_m = 0
Ve_Fe = [(Ve_m * valor) for valor in Fe[:-1]] + [380]

print(Ve_Fe)
```

```python
Ns_new1 = []
for valor in Fe:
    Ns_new = (valor * 60) / 4
    Ns_new1.append(Ns_new)

print(Ns_new1)
```

```python
Rs = 0.5907
Xs_new1 = []
for valor2 in Fe:
  Xs_new = (2*np.pi*valor2)*(0.0009007)
  Xs_new1.append(Xs_new)
print(Xs_new1)
Xm_new1 = []
for valor2 in Fe:
  Xm_new = (2*np.pi*valor2)*(0.03498)
  Xm_new1.append(Xm_new)
print(Xm_new1)
R1 = 0.3492
X1_new1 = []
for valor2 in Fe:
  X1_new = (2*np.pi*valor2)*(0.004182)
  X1_new1.append(X1_new)
print(X1_new1)
R2 = 0.3503
X2_new1 = []
for valor2 in Fe:
  X2_new = (2*np.pi*valor2)*(0.0009007)
  X2_new1.append(X2_new)
print(X2_new1)
```

# **<font color="blue"><center>Para 21 Hz con Ve en 133 Vrms y 315 RPM**

```python
Sn0 = (900 - 315 ) / 900

Rr_01 = (R1*R2*(R1+R2)+Sn0**2*(R1*X2_new1[0]**2+R2*X1_new1[0]**2)) / ((R1+R2)**2+Sn0**2*(X1_new1[0]+X2_new1[0])**2)
Xr_01= (R1**2 * X2_new1[0] + R2**2 * X1_new1[0] + Sn0**2 * X1_new1[0] * X2_new1[0] * (X1_new1[0] + X2_new1[0])) / ((R1 + R2)**2 + Sn0**2 * (X1_new1[0] + X2_new1[0])**2)

pp=(1j*Xm_new1[0]*((Rr_01)+1j*Xr_01))/(1j*Xm_new1[0]+((Rr_01)+1j*Xr_01))
Zeq_01=Rs+1j*Xs_new1[0]+(pp)
r01, theta01 = abs(Zeq_01), cmath.phase(Zeq_01)


Ie_01 = Ve_Fe[0]/r01
print(Ie_01)

Vth_50hz = Ve_Fe[0]*(((Xm_new1[0])/((Rs**2)+((Xs_new1[0]+Xm_new1[0])**2))**0.5))
Zth_50 = (((complex(0, Xm_new1[0]))*(Rs + complex(0, Xs_new1[0]))/(Rs + (complex(0, Xs_new1[0])+ complex(0, Xm_new1[0])))))
Rth_50 = Zth_50.real
Xth_50 = Zth_50.imag
Ws_50hz = ((2*np.pi)/60)*Ns_new1[0]
```

# **<font color="blue"><center>Para 40 Hz con Ve en 253.33 Vrms y 600 RPM**

```python
Sn02 = (900 - 600) / 900

Rr_02 = (R1*R2*(R1+R2)+Sn02**2*(R1*X2_new1[1]**2+R2*X1_new1[1]**2)) / ((R1+R2)**2+Sn02**2*(X1_new1[1]+X2_new1[1])**2)
Xr_02= (R1**2 * X2_new1[1] + R2**2 * X1_new1[1] + Sn02**2 * X1_new1[1] * X2_new1[1] * (X1_new1[1] + X2_new1[1])) / ((R1 + R2)**2 + Sn02**2 * (X1_new1[1] + X2_new1[1])**2)

pp=(1j*Xm_new1[1]*((Rr_02)+1j*Xr_02))/(1j*Xm_new1[1]+((Rr_02)+1j*Xr_02))
Zeq_02=Rs+1j*Xs_new1[1]+(pp)
r02, theta02 = abs(Zeq_02), cmath.phase(Zeq_02)


Ie_02 = Ve_Fe[1]/r02
print(Ie_02)

Vth_20hz = Ve_Fe[1]*(((Xm_new1[1])/((Rs**2)+((Xs_new1[1]+Xm_new1[1])**2))**0.5))
Zth_20 = (((complex(0, Xm_new1[1]))*(Rs + complex(0, Xs_new1[1]))/(Rs + (complex(0, Xs_new1[1])+ complex(0, Xm_new1[1])))))
Rth_20 = Zth_20.real
Xth_20 = Zth_20.imag
Ws_20hz = ((2*np.pi)/60)*Ns_new1[1]
```

```python
Sn03 = (900 - 1080) / 900

Rr_03 = (R1*R2*(R1+R2)+Sn03**2*(R1*X2_new1[2]**2+R2*X1_new1[2]**2)) / ((R1+R2)**2+Sn03**2*(X1_new1[2]+X2_new1[2])**2)
Xr_03= (R1**2 * X2_new1[2] + R2**2 * X1_new1[2] + Sn03**2 * X1_new1[2] * X2_new1[2] * (X1_new1[1] + X2_new1[2])) / ((R1 + R2)**2 + Sn03**2 * (X1_new1[2] + X2_new1[2])**2)

pp=(1j*Xm_new1[2]*((Rr_03)+1j*Xr_03))/(1j*Xm_new1[2]+((Rr_03)+1j*Xr_03))
Zeq_03=Rs+1j*Xs_new1[2]+(pp)
r03, theta03 = abs(Zeq_03), cmath.phase(Zeq_03)


Ie_03 = Ve_Fe[2]/r03
print(Ie_03)

Vth_75hz = Ve_Fe[2]*(((Xm_new1[2])/((Rs**2)+((Xs_new1[2]+Xm_new1[2])**2))**0.5))
Zth_75 = (((complex(0, Xm_new1[2]))*(Rs + complex(0, Xs_new1[2]))/(Rs + (complex(0, Xs_new1[2])+ complex(0, Xm_new1[2])))))
Rth_75 = Zth_75.real
Xth_75 = Zth_75.imag
Ws_75hz =  ((2*np.pi)/60)*Ns_new1[2]
```
